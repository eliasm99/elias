```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from shapely.geometry import Point, Polygon
from pysal.explore.esda.moran import Moran, Moran_Local
from pysal.lib import weights
from libpysal.weights import Queen, Rook, DistanceBand
from sklearn.preprocessing import MinMaxScaler
import contextily as ctx
import warnings
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from matplotlib.colors import Normalize,ListedColormap
import networkx as nx
from geopy.distance import geodesic
from scipy.spatial.distance import pdist, squareform
import random
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from folium import Map, Circle, LayerControl
from folium.plugins import HeatMap
import matplotlib.cm as cm
import matplotlib.colors as mcolors
from matplotlib.collections import LineCollection
```

```python
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Preprocessed_Dataset.csv')
```

```python
# Step 1: Standardize Pollutant Data
pollutant_columns = ['SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']
scaler = StandardScaler()
data[pollutant_columns] = scaler.fit_transform(data[pollutant_columns])
```

```python
# Extract station codes and coordinates
coordinates = data[['Station code', 'Latitude', 'Longitude']].drop_duplicates()
coordinates.set_index('Station code', inplace=True)
```

```python
# Compute pairwise distances using station codes as indices
distances = pdist(coordinates[['Latitude', 'Longitude']], metric='euclidean')
distance_matrix = squareform(distances)
```

```python
# Create a DataFrame for the distance matrix with station codes as index and columns
distance_df = pd.DataFrame(
    distance_matrix,
    index=coordinates.index,
    columns=coordinates.index
)
```

```python
distance_df.head()
```

```python
station_averages = data.groupby('Station code')[pollutant_columns].mean()
```

```python
# Step 2: Create a correlation plot for pollutants
plt.figure(figsize=(10, 8))
sns.pairplot(station_averages, diag_kind="kde", corner=True)
plt.suptitle("Correlation between Pollutants", y=1.02)
plt.show()
```

```python
# Pairwise differences for all pollutants
pairwise_differences = {}
for pollutant in pollutant_columns:
    diff_matrix = pd.DataFrame(
        np.abs(station_averages[pollutant].values[:, None] - station_averages[pollutant].values),
        index=station_averages.index,
        columns=station_averages.index
    )
    pairwise_differences[pollutant] = diff_matrix.stack().reset_index(drop=True)
```

```python
# Flatten the distance matrix to align with pollutant differences
distances = distance_df.stack().reset_index(drop=True)
```

```python
# Combine distances and pollutant differences into a single DataFrame
combined_data = pd.DataFrame({
    'Distance': distances,
    **{f'{pollutant}_diff': pairwise_differences[pollutant] for pollutant in pollutant_columns}
})
```

```python
# Step 4: Plot each pollutant's relationship with spatial proximity
plt.figure(figsize=(18, 12))
for i, pollutant in enumerate(pollutant_columns, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(
        data=combined_data,
        x='Distance',
        y=f'{pollutant}_diff',
        alpha=0.6
    )
    plt.title(f'Relationship Between Distance and {pollutant} Differences')
    plt.xlabel('Distance (km)')
    plt.ylabel(f'{pollutant} Difference')
    plt.grid()

plt.tight_layout()
```

```python
# Step 1: Aggregate pollutant levels per station
pollutant_columns = ['SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']
station_aggregates = data.groupby('Station code')[pollutant_columns + ['Latitude', 'Longitude']].mean()
```

```python
# Convert to GeoDataFrame
geo_data = gpd.GeoDataFrame(
    station_aggregates,
    geometry=gpd.points_from_xy(station_aggregates.Longitude, station_aggregates.Latitude)
)
```

```python
# Step 2: Compute spatial weights (Queen contiguity)
weights = Queen.from_dataframe(geo_data)
```

```python
# Step 3: Calculate Moran's I for each pollutant
moran_results = {}
for pollutant in pollutant_columns:
    moran = Moran(geo_data[pollutant], weights)
    moran_results[pollutant] = {
        "Moran's I": moran.I,
        "p-value": moran.p_sim,
        "z-score": moran.z_sim
    }
```

```python
# Display Moran's I results
moran_df = pd.DataFrame(moran_results).T
print("Moran's I Results:\n", moran_df)
```

```python
# Step 4: Perform Local Moran's I (LISA) for each pollutant and plot
for pollutant in pollutant_columns:
    moran_local = Moran_Local(geo_data[pollutant], weights)
    geo_data[f'{pollutant}_LISA'] = moran_local.Is
    geo_data[f'{pollutant}_LISA_sig'] = moran_local.p_sim < 0.05

    # Plot significant clusters
    plt.figure(figsize=(10, 8))
    geo_data.plot(
        column=f'{pollutant}_LISA',
        cmap='coolwarm',
        legend=True,
        scheme='quantiles',
        alpha=0.8
    )
    plt.title(f"Local Moran's I (LISA) for {pollutant}")
    plt.show()
```

```python
# Define colormap for LISA significance categories
cmap = ListedColormap(['blue', 'lightblue', 'pink', 'red'])

# Step 4: Plot Spatial Autocorrelation Maps for each pollutant
pollutant_columns = ['SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']
```

```python
for pollutant in pollutant_columns:
    lisa_column = f'{pollutant}_LISA'
    sig_column = f'{pollutant}_LISA_sig'
    
    # Create a GeoDataFrame for significant LISA values
    significant_geo_data = geo_data[geo_data[sig_column]]

    plt.figure(figsize=(12, 10))
    
    # Plot all stations with LISA values
    base = geo_data.plot(
        column=lisa_column,
        cmap='coolwarm',
        legend=True,
        alpha=0.5,
        figsize=(12, 8)
    )
    
    # Overlay significant clusters
    significant_geo_data.plot(
        color='black',
        markersize=50,
        alpha=0.7,
        ax=base
    )
    
    # Add station code labels
    for x, y, label in zip(geo_data.geometry.x, geo_data.geometry.y, geo_data.index):
        plt.text(x, y, label, fontsize=8, ha='right', va='bottom', color='darkblue')
    
    # Customize plot
    plt.title(f"Local Moran's I (LISA) for {pollutant} with Significant Clusters")
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.grid()
    plt.show()
```

```python
# Step 1: Filter Data for a Specific Time Period
time_filter = '2017-01'  # Example: January 2017
filtered_data = data[data['Measurement date'].str.startswith(time_filter)]
```

```python
# Step 2: Aggregate Data by Station
pollutant_columns = ['SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']
station_averages = filtered_data.groupby('Station code')[pollutant_columns].mean()
```

```python
# Step 3: Build a Weighted Graph for a Pollutant
pollutant = 'PM2.5'  # Focus on one pollutant for visualization
graph = nx.Graph()
```

```python
# Add nodes with station information
for station, row in station_averages.iterrows():
    graph.add_node(station, value=row[pollutant])
```

```python
# Add edges based on spatial proximity
threshold_distance = 5.0  # Example: 5 km threshold for edge creation
for station1 in distance_df.index:
    for station2 in distance_df.columns:
        if station1 != station2:  # Avoid self-loops
            distance = distance_df.loc[station1, station2]
            if distance <= threshold_distance:
                # Ensure station exists in station_averages
                if station1 in station_averages.index and station2 in station_averages.index:
                    weight = abs(station_averages.loc[station1, pollutant] -
                                 station_averages.loc[station2, pollutant]) / distance
                    graph.add_edge(station1, station2, weight=weight)
```

```python
# Step 4: Visualize the Network
plt.figure(figsize=(12, 10))

# Position nodes using a spring layout
pos = nx.spring_layout(graph, seed=42)

# Draw nodes
node_colors = [graph.nodes[node]['value'] for node in graph.nodes]
nx.draw_networkx_nodes(
    graph,
    pos,
    node_size=500,
    cmap=plt.cm.Blues,
    node_color=node_colors
)

# Prepare edge weights for color mapping
edge_weights = [d['weight'] for (_, _, d) in graph.edges(data=True)]
edge_cmap = cm.coolwarm
edge_norm = mcolors.Normalize(vmin=min(edge_weights), vmax=max(edge_weights))

# Draw edges with the colormap
nx.draw_networkx_edges(
    graph,
    pos,
    alpha=0.5,
    edge_color=edge_weights,
    edge_cmap=edge_cmap,
    edge_vmin=min(edge_weights),
    edge_vmax=max(edge_weights)
)

# Draw labels
nx.draw_networkx_labels(graph, pos, font_size=8, font_color="black")

# Add colorbar for edges
sm = cm.ScalarMappable(cmap=edge_cmap, norm=edge_norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=plt.gca())
cbar.set_label("Edge Weight (Pollutant Difference / Distance)")

# Customize and show plot
plt.title(f"Dynamic Network Visualization for {pollutant} ({time_filter})")
plt.show()
```

```python
# Step 1: Filter Data for Multiple Time Periods
time_periods = ['2017-01', '2017-02', '2017-03','2017-04', '2017-05', '2017-06']  # Example: Three months
filtered_data = data[data['Measurement date'].str.slice(0, 7).isin(time_periods)]
```

```python
# Step 2: Aggregate Pollutant Data by Station and Time Period
pollutant = 'PM2.5'  # Focus on a single pollutant for visualization
filtered_data['Time Period'] = filtered_data['Measurement date'].str.slice(0, 7)
grouped_data = filtered_data.groupby(['Station code', 'Time Period'])[pollutant].mean().unstack()
```

```python
# Step 3: Create Line Segments for Storyline
stations = grouped_data.index
time_steps = grouped_data.columns
values = grouped_data.values
```

```python
# Normalize pollutant values for better visualization
norm_values = (values - np.nanmin(values)) / (np.nanmax(values) - np.nanmin(values))

# Generate coordinates for the storyline
y_positions = np.arange(len(stations))  # One row per station
x_positions = np.arange(len(time_steps))  # Columns for each time step
```

```python
# Prepare line segments
line_segments = []
for i, station in enumerate(stations):
    for j in range(len(time_steps) - 1):
        line_segments.append([
            (x_positions[j], y_positions[i] + norm_values[i, j]),
            (x_positions[j + 1], y_positions[i] + norm_values[i, j + 1])
        ])
```

```python
# Step 4: Plot the Storyline Visualization
fig, ax = plt.subplots(figsize=(12, 8))

# Add line segments
line_collection = LineCollection(
    line_segments,
    cmap='coolwarm',
    linewidths=2,
    norm=plt.Normalize(vmin=np.nanmin(values), vmax=np.nanmax(values))
)
line_collection.set_array(norm_values.flatten())
ax.add_collection(line_collection)

# Customize plot
ax.autoscale()
ax.set_xticks(x_positions)
ax.set_xticklabels(time_steps, rotation=45)
ax.set_yticks(y_positions)
ax.set_yticklabels(stations)
ax.set_title(f"Storyline Visualization of {pollutant} Diffusion Over Time")
ax.set_xlabel("Time Period")
ax.set_ylabel("Stations")
cbar = plt.colorbar(line_collection, ax=ax)
cbar.set_label(f"Normalized {pollutant} Levels")

plt.tight_layout()
```
